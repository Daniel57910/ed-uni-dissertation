```python
import cudf as gpu_pd
import pandas as pd

METADATA_STAT_COLUMNS = [
    'user_id',
    'session_30_raw',
    'cum_session_time_raw',
    'date_time',
]
```

```python
df = gpu_pd.read_parquet('rl_ready_data/files_used_30/rl_ready_data.parquet')
df_original = gpu_pd.read_parquet('calculated_features/files_used_30.parquet')
print(df_original.isna().sum())
```

```python
exp_df = df[METADATA_STAT_COLUMNS]
sample_session = exp_df[exp_df['session_30_raw'] > 1].sample(1)
sample_user, sample_session = sample_session['user_id'].values[0].item(), sample_session['session_30_raw'].values[0].item()
sample_df = exp_df[(exp_df['user_id'] == sample_user) & (exp_df['session_30_raw'] == sample_session)]
sample_df = sample_df.sort_values(by=['date_time'])
selection = sample_df[['user_id', 'session_30_raw', 'cum_session_time_raw']].groupby(['user_id', 'session_30_raw']).agg({'cum_session_time_raw': ['max']})
```
